```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Player': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Goals': [5, 7, 3, 8, 6],
    'Assists': [2, 3, 5, 1, 4]
}

df = pd.DataFrame(data)

# Display basic stats
print(df.describe())

# Plot Goals
plt.bar(df['Player'], df['Goals'], color='skyblue')
plt.title('Goals by Player')
plt.xlabel('Player')
plt.ylabel('Goals')
plt.show()
```
